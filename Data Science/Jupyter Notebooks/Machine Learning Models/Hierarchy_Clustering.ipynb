## Hierarchy Clustering
In data mining and statistics, hierarchical clustering is a method of cluster analysis which seeks to build a hierarchy of clusters.

![title](images/Hierarchy_Clustering.jpg)

## Prepare Data

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# get data
iris = load_iris()
x = iris.data
y = iris.target

# split data into train(80%) & test(20%) data since we don't have anything to test on
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

## Hierarchy Clustering Classifier

```python
import scipy.cluster.hierarchy as HierarchyClustering
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(10, 7))  
plt.title("Dendrograms")  
dend = HierarchyClustering.dendrogram(HierarchyClustering.linkage(x, method='ward'))
plt.axhline(y=6, color='r', linestyle='--')
```
