## K-Means Clustering
K-means algorithm identifies k number of centroids, and then allocates every data point to the nearest cluster, while keeping the centroids as small as possible.

![title](images/K_Means_Clustering.png)

## Prepare Data

```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set() 
from sklearn.datasets.samples_generator import make_blobs
%matplotlib inline

# create random clusters
x, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
plt.scatter(x[:, 0], x[:, 1], s=50);
```

## K-Means Clustering Classifier

```python
from sklearn.cluster import KMeans
from sklearn import metrics

# initialize classifier
kMeans = KMeans(n_clusters=4)

# train/fit the model
kMeans.fit(x, y)

# use the model to predict the classes of the test data
y_predict = kMeans.predict(x)

# get the accuracy score
metrics.accuracy_score(y, y_predict)
```

```python
plt.scatter(x[:, 0], x[:, 1], c=y_predict, s=50, cmap='viridis')

centers = kMeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.5)
```
